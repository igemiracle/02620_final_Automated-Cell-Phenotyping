```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Heatmap

```python
# plot the heatmap
datasets_F1 = {
    'Baron': [0.495, 0.502, 0.632, 0.611],
    'CellBench': [0.635, 0.997, 0.978, 0.996],
    'Zheng': [0.318, 0.920, 0.668, 0.856]
}

# Create DataFrame
heatmap_df = pd.DataFrame(datasets_F1, index=['K-means', 'SVM', 'Logistic Regression', 'XGBoost'])

# Creating the heatmap
plt.figure(figsize=(7, 6))
ax = sns.heatmap(heatmap_df, annot=True, fmt='.3f', cmap='YlOrRd', cbar_kws={'label': 'F1-Score'})
ax.set_yticklabels(ax.get_yticklabels(), rotation=0)
plt.title('F1-Score Heatmap', fontsize=14)
plt.xlabel('Datasets', fontsize=12)
plt.ylabel('Methods', fontsize=12)
plt.show()
```

## Bar chart

```python
# plot the bar chart for train error and test error

error = {
    'Dataset': ['Baron', 'Baron', 'Baron', 'Baron', 'Baron', 'Baron', 
                'CellBench', 'CellBench', 'CellBench', 'CellBench', 'CellBench', 'CellBench', 
                'Zheng', 'Zheng', 'Zheng', 'Zheng', 'Zheng', 'Zheng'],
    'Error': ['LogRegre Train Error', 'LogRegre Test Error', 'SVM Train Error', 'SVM Test Error', 'XGBoost Train Error', 'XGBoost Test Error', 
               'LogRegre Train Error', 'LogRegre Test Error', 'SVM Train Error', 'SVM Test Error', 'XGBoost Train Error', 'XGBoost Test Error',
               'LogRegre Train Error', 'LogRegre Test Error', 'SVM Train Error', 'SVM Test Error', 'XGBoost Train Error', 'XGBoost Test Error'],
    'values': [0.330059, 0.3302674, 0.08979592802470161, 0.07345808879184441, 0.0311086797957695, 0.07221006564551424, 
               0.021952, 0.0220254, 0.0, 0.0, 0.0, 0.0029583657419410825, 
               0.3069027, 0.3083125, 0.09768055555555553, 0.07024999999999999, 0.062671875, 0.1434375]
}

colors = ['#db3124', '#fc8c5a', '#ffdf92', '#e6f1f3', '#90bee0', '#4b74b2']

error_df = pd.DataFrame(error)

# Pivot the data for plotting
df_error_pivot = error_df.pivot(index='Dataset', columns='Error', values='values')

# Plot
ax = df_error_pivot.plot(kind='bar', figsize=(10, 6), width=0.7, color=colors)
plt.title('Training and Test Errors for Logistic Regression, SVM and XGBoost', fontsize=14)
plt.ylabel('Error', fontsize=12)
plt.xlabel('Datasets', fontsize=12)
plt.xticks(rotation=0)
plt.legend(bbox_to_anchor=(1, 1))
plt.tight_layout()

plt.show()
```
